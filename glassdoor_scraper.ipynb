```python
#import necessary libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from shutil import which
from selenium.webdriver.edge.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.edge.options import Options
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException, ElementNotInteractableException
import pandas as pd
import time
```

```python
def fetch_jobs(keyword, num_pages):
    options = Options()
    options.add_argument("window-size=1920,1080")
    options.add_argument("--incognito")
    options.add_argument("--disable-site-isolation-trials")
    #Enter your chromedriver.exe path below
    time.sleep(2)
    driver = webdriver.Edge(service=Service("C:/Users/sunne/Downloads/msedgedriver.exe"), options=options)
    driver.get("https://www.glassdoor.com/Job/nigeria-jobs-SRCH_IL.0,7_IN177.htm")
    search_input = driver.find_element(By.ID,"sc.keyword")
    search_input.send_keys(keyword)
    search_input.send_keys(Keys.ENTER)
    time.sleep(2)
                          
    #driver.find_element_by_xpath('//*[@id="Discover"]/div/div/div[1]/div[1]/div[3]/a').click()
    
    company_name = []
    job_title = []
    salary_est = []
    location = []
    job_description = []
    salary_estimate = []
    company_size = []
    company_type = []
    company_sector = []
    company_industry = []
    company_founded = []
    company_revenue = []
    
    
    
    #Set current page to 1
    current_page = 1     
            
    time.sleep(3)
    
    while current_page <= num_pages:   
        
        done = False
        while not done:
            job_cards = driver.find_elements(By.XPATH,"//article[@id='MainCol']//ul/li[@data-adv-type='GENERAL']")
            for card in job_cards:
                card.click()
                time.sleep(2)

                #Closes the signup prompt
                try:
                    driver.find_element(By.XPATH,'//*[@id="LoginModal"]/div/div/div[2]/button').click()
                    time.sleep(5)
                except NoSuchElementException:
                    time.sleep(2)                          
                    pass
                except ElementClickInterceptedException:
                    driver.find_element(By.XPATH,'//*[@id="LoginModal"]/div/div/div[2]/button').click()
                    time.sleep(5)

                #Expands the Description section by clicking on Show More
                try:
                    driver.find_element(By.XPATH,'//*[@id="JobDescriptionContainer"]/div[2]').click()
                    time.sleep(1)
                except NoSuchElementException:
                    card.click()
                    print(str(current_page) + '#ERROR: no such element')
                    time.sleep(30)
                    driver.find_element(By.XPATH,'//*[@id="JobDescriptionContainer"]/div[2]').click()
                except ElementNotInteractableException:
                    card.click()
                    driver.implicitly_wait(30)
                    print(str(current_page) + '#ERROR: not interactable')
                    driver.find_element(By.XPATH,'//*[@id="JobDescriptionContainer"]/div[2]').click()

                #Scrape 

                try:
                    company_name.append(driver.find_element(By.XPATH,'//*[@id="JDCol"]/div/article/div/div[1]/div/div/div[1]/div[3]/div[1]/div[1]/div').text)
                except:
                    company_name.append("#N/A")
                    pass
                
                try:
                    job_title.append(driver.find_element(By.XPATH,'//*[@id="JDCol"]/div/article/div/div[1]/div/div/div[1]/div[3]/div[1]/div[2]').text)
                except:
                    job_title.append("#N/A")
                    pass

                try:
                    location.append(driver.find_element(By.XPATH,'//*[@id="JDCol"]/div/article/div/div[1]/div/div/div[1]/div[3]/div[1]/div[3]').text)
                except:
                    location.append("#N/A")
                    pass

                try:
                    job_description.append(driver.find_element(By.XPATH,'//*[@id="JobDescriptionContainer"]').text)
                except:
                    job_description.append("#N/A")
                    pass

                try:
                    salary_estimate.append(driver.find_element(By.XPATH,'//*[@id="JDCol"]/div/article/div/div[1]/div/div/div[1]/div[3]/div[1]/div[4]/span').text)
                except:
                    salary_estimate.append("#N/A")
                    pass
                
                done = True
                
       # Moves to the next page         
        if done:
            print(str(current_page) + ' ' + 'out of' +' '+ str(num_pages) + ' ' + 'pages done')
            driver.find_element(By.XPATH, '//*[@id="MainCol"]/div[2]/div/div[1]/button[5]').click()   
            current_page = current_page + 1
            time.sleep(4)
            




    driver.close()
    df = pd.DataFrame({'company': company_name,
                       'job title': job_title,
                       'location': location,
                       'job description': job_description,
                       'salary estimate': salary_estimate})
    df.to_csv(keyword + '.csv')
```
